```python
emailDir = "./maildir/maildir"
rawcsv = "../experiments/raw.csv"
metadataHeaders = '../experiments/metadataHeaders.csv'
```

```python
import pandas as pd

columns = pd.read_csv(metadataHeaders, sep=',').columns.tolist()
columns
```

```python
df = pd.read_csv(rawcsv, names=columns, sep='|', low_memory=False)

print("-- DONE --")
```

```python
import hashlib

def hashFunc(val):
    try:
        return hashlib.sha256(val.encode('utf-8')).hexdigest()
    except Exception:
        pass

df.Person = df.Person.apply(hashFunc)
df.Directory = df.Directory.apply(hashFunc)
df.To = df.To.apply(hashFunc)
df.From = df.From.apply(hashFunc)
df['X-To'] = df['X-To'].apply(hashFunc)
df['X-From'] = df['X-From'].apply(hashFunc)
df.Cc = df.Cc.apply(hashFunc)
df.Bcc = df.Bcc.apply(hashFunc)
df['X-cc'] = df['X-cc'].apply(hashFunc)
df['X-bcc'] = df['X-From'].apply(hashFunc)
df['X-Folder'] = df['X-Folder'].apply(hashFunc)
df['X-Origin'] = df['X-Origin'].apply(hashFunc)
df['X-FileName'] = df['X-FileName'].apply(hashFunc)

df.head()
```

```python
df.to_csv('../experiments/anon.csv', sep='|', header=False)
print("-- DONE --")
```
